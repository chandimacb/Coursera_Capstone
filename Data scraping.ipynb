Adding the URL containing the dataset to url

```python
url = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
html = urlopen(url)
```

Creating a Beautiful Soup Object from the HTML file

```python
soup = BeautifulSoup(html, 'html5lib')
type(soup)
```

Extraxting the title of the webpage to check that the correct web page is loaded

```python
title = soup.title
print(title)
```

Getting all the rows in the dataset in to a list form

```python
rows = soup.find_all('tr')
for row in rows:
    row_td = row.find_all('td')
print(row_td)
type(row_td)
```

Removing HTML tags

```python
import re

list_rows = []
for row in rows:
    cells = row.find_all('td')
    str_cells = str(cells)
    clean = re.compile('<.*?>')
    clean2 = (re.sub(clean, '',str_cells))
    list_rows.append(clean2)
print(clean2)
type(clean2)
```

converting the list into a dataframe using pandas

```python
df = pd.DataFrame(list_rows)
df.head(181)
```

splitting the columns
